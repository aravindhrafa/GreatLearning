# Hypothesis testing

# Is saving money(finances) gender dependent?

Is saving money(finances) gender dependent

```python
male_finance = dataframe2.loc[dataframe2['Gender']== 'male','Finances']
male_finance
```

```python
male_finance.count()
```

```python
female_finance = dataframe2.loc[dataframe2['Gender']== 'female','Finances']
female_finance
```

```python
female_finance.count()
```

```python
np.warnings.filterwarnings('ignore')
```

```python
from scipy.stats import ttest_ind
t_test,p_val = ttest_ind(female_finance,male_finance,equal_var= False)
```

```python
t_test
```

```python
p_val
```

p-value = 0.06004 > critical value (0.05). Hence failed to reject the null hypothesis.Hence Saving money is not Gender dependent

```python
city_finance = dataframe2.loc[dataframe2['Village - town' ]== 'city','Finances']
city_finance
```

```python
city_finance.count()
```

```python
village_finance =dataframe2.loc[dataframe2['Village - town'] == 'village','Finances']
village_finance
```

```python
village_finance.count()
```

```python
ttest_ind(city_finance,village_finance)
```

The p-value = 0.008034328 which is found to be lower than 0.05. Hence rejecting the null hypothesis
Based on Hypothesis, there is difference is people with their finances in city as compared to the village

# Relation between gender and lifestyle

```python
men_lifestyle = dataframe2.loc[dataframe2['Gender']=='male','Healthy eating']
men_lifestyle
```

```python
women_lifestyle = dataframe2.loc[dataframe2['Gender']=='female','Healthy eating']
women_lifestyle
```

```python
ttest_ind(men_lifestyle,women_lifestyle)
```

Since the p-value=0.04704767254, which is less than critical value(0.05), hence rejecting the Null hypothesis.
Hence, women are found to be healthier than the men
