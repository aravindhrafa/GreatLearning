### Generate tf-idf matrix using the column `description`. Consider till 3-grams, with minimum document frequency as 0.

```python

tf_idf_vect = TfidfVectorizer(analyzer='word',ngram_range=(1,3),stop_words='english', min_df = 0)

# Another form of vector like CountVect
tf_idf_vect.fit(dataframe1['description']) # Values are being fitted here 
description_matrix = tf_idf_vect.transform(dataframe1['description'])  # Matrix creation
```

```python
#title_matrix

print(description_matrix)
```

### create cosine similarity matrix

```python
# cosine_matrix_for_description = cosine_similarity(description_matrix)

from sklearn.metrics.pairwise import linear_kernel

cosine_similarity_matrix = linear_kernel(description_matrix, description_matrix)
```
