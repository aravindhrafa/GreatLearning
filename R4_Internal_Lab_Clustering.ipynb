### K - Means Clustering

### 4. Assign a dummy array called distortion

```python
distortion = []
```

You can later use this array to plot the elbow plot

```python
from sklearn.cluster import KMeans
from scipy.spatial import distance
```

### 5. Calculating distortion - C distance

```python
cluster_error = []
cluster_silh_score = []
cluster_range = range(1,10)
```

Iterating values of k from 1 to 10 fit K means model
Using c distance - Get the measure for Sum of squares error.

```python
for num_cluster in cluster_range:
    clusters= KMeans(num_cluster,n_init=5)
    clusters.fit(dataset3)
    labels = clusters.labels_
    centroids = clusters.cluster_centers_   #The value that is arrived from the Kmeans Algorithm
    cluster_error.append(clusters.inertia_)   # The error or inertia values are stored on the cluster _errors array
    distortion.append(sum(np.min(distance.cdist(dataset3,clusters.cluster_centers_,'euclidean'),axis=1))/dataset3.shape[0])


print(len(cluster_range))
print(len(cluster_error))
    
```

```python
    
clusters_dataframe = pd.DataFrame({"num_clusters":cluster_range,"clusters_errors":cluster_error})
clusters_dataframe
```

```python
plt.figure(figsize = (12,6))
plt.plot(clusters_dataframe.num_clusters,clusters_dataframe.clusters_errors,marker='o')
```

### 6. Plotting Elbow/ Scree Plot

```python
plt.figure(figsize = (12,6))
plt.plot(clusters_dataframe.num_clusters,clusters_dataframe.clusters_errors,marker='o')
```

Use Matplotlib to plot the scree plot - Note: Scree plot plots distortion vs the no of clusters

```python
plt.plot(clusters_dataframe.num_clusters,distortion,marker='o')
```

## Hierarchical Clustering 

### 8. Variable creation

For Hierarchical clustering, we will create datasets using multivariate normal distribution to visually observe how the clusters are formed at the end

```python
a = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]], size=[100,])
b = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], size=[50,])
c = np.random.multivariate_normal([10, 20], [[3, 1], [1, 4]], size=[100,])
```

### 9. Combine all three arrays a,b,c into a dataframe

```python
dataframe6= np.concatenate([a,b,c])
```

```python
dataframe7=pd.DataFrame(dataframe6)
```

```python
type(dataframe7)
```

### 10. Use scatter matrix to print all the 3 distributions

```python
sns.pairplot(dataframe7)
```

### 11. Find out the linkage matrix

```python
from scipy.cluster.hierarchy import cophenet, dendrogram, linkage
from scipy.spatial.distance import pdist
```

Use ward as linkage metric and distance as Eucledian

```python
Z = linkage(dataframe7, 'ward')
c, coph_dists = cophenet(Z , pdist(dataframe7))
c
```

```python
Z.shape
```

```python
Z[1:20]
```

### 12. Plot the dendrogram for the consolidated dataframe

```python
plt.figure(figsize=(10, 10))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(Z, leaf_rotation=90.,color_threshold = 30, leaf_font_size=8. )
plt.tight_layout()
```

### 13. Recreate the dendrogram for last 12 merged clusters 

```python
dendrogram(Z, 12, truncate_mode='lastp')
```

Hint: Use truncate_mode='lastp' attribute in dendrogram function to arrive at dendrogram 

### 14. From the truncated dendrogram, find out the optimal distance between clusters which u want to use an input for clustering data

```python
maxd=50
```

### 15. Using this distance measure and fcluster function to cluster the data into 3 different groups

```python
from scipy.cluster.hierarchy import fcluster
```

```python
cluster = fcluster(Z, maxd, criterion='distance')
cluster
```
